```
jam_hazfi = pd.read_csv('Jam-e-hazfi.csv')
pd.set_option("display.max_rows", None)
display(jam_hazfi)
```

```
todays_date = datetime.datetime.now().date()
index = np.array([np.arange(len(jam_hazfi.index))]).flatten()

columns = ['Type', 'Year', 'Stage', 'Team Right', 'Team Left', 'Score Right', 'Score Left', 'Time', 'Date', 'Penalty Right', 'Penalty Left']
df_jam_hazfi = pd.DataFrame(index=index, columns=columns)

#pd.date_range(todays_date-datetime.timedelta(10), periods=10, freq='D')
```

```
print(soup_jam_hazfi.find_all(class_='team-name left')[0].get_text())
print(soup_jam_hazfi.find_all(class_='team-name right')[0].get_text())
```

```
for i in range(len(jam_hazfi.index)):
    soup_temp = BeautifulSoup(jam_hazfi.html[i], 'html.parser')
    team_left = soup_temp.find_all(class_='team-name left')[0].get_text()
    team_right = soup_temp.find_all(class_='team-name right')[0].get_text()
    score_left = soup_temp.find_all(class_='team-score left')[0].get_text()
    score_right = soup_temp.find_all(class_='team-score right')[0].get_text()
    time = soup_temp.find_all(class_='match-time pull-right')[0].get_text()
    date = soup_temp.find_all(class_='match-date pull-right')[0].get_text()
    df_jam_hazfi.at[i, 'Team Left'] = team_left
    df_jam_hazfi.at[i, 'Team Right'] = team_right
    
    df_jam_hazfi.at[i, 'Score Left'] = score_left[-1]
    df_jam_hazfi.at[i, 'Score Right'] = score_right[0]

    df_jam_hazfi.at[i, 'Stage'] = jam_hazfi.Stage[i]
    df_jam_hazfi.at[i, 'Year'] = jam_hazfi.Year[i][15:20]
    df_jam_hazfi.at[i, 'Type'] = jam_hazfi.Year[i][0:15]
    df_jam_hazfi.at[i, 'Time'] = time
    df_jam_hazfi.at[i, 'Date'] = date

    start = score_left.find("(") + len("(")
    # print(start)
    end = score_left.find(")")
    penalty_left = score_left[start:end]
    if start is 0:
        df_jam_hazfi.at[i, 'Penalty Left'] = '0'
    else:
        df_jam_hazfi.at[i, 'Penalty Left'] = penalty_left

    start = score_right.find("(") + len("(")
    end = score_right.find(")")
    penalty_right = score_right[start:end]
    if start is 0:
        df_jam_hazfi.at[i, 'Penalty Right'] = '0'
    else:
        df_jam_hazfi.at[i, 'Penalty Right'] = penalty_right
display(df_jam_hazfi)
print(df_jam_hazfi.isna().sum())
df_jam_hazfi.to_csv("jam_hazfi_data.csv", index=False)
```

```
def plot_hist(data, name, size=(10, 7)):
    legend = ['distribution'] 
    fig, axs = plt.subplots(1, 1, 
                            figsize=size,  
                            tight_layout = True)  
        
    # Add padding between axes and labels  
    axs.xaxis.set_tick_params(pad=5)  
    axs.yaxis.set_tick_params(pad=10)  
    # Add x, y gridlines  
    axs.grid(b=True, color='grey', linestyle='-.', linewidth=0.5, alpha=0.6)

    n_bins = (len(data.value_counts()))
    print(data.value_counts())#.index)
    N, bins, patches = axs.hist(data, bins=int(n_bins), label='')
    plt.draw()
    ticks = [(t.get_text()) for t in axs.get_xticklabels()]
    # Setting color 
    fracs = ((N**(1 / 5)) / N.max()) 
    norm = colors.Normalize(fracs.min(), fracs.max()) 
    
    for thisfrac, thispatch in zip(fracs, patches): 
        color = plt.cm.viridis(norm(thisfrac)) 
        thispatch.set_facecolor(color)

    # Label the raw counts and the percentages below the x-axis...
    bin_centers = 0.5 * np.diff(bins) + bins[:-1]
    for count, x in zip(ticks, bin_centers):
        # Label the raw counts
        axs.annotate(int(count), xy=(x, 0), xycoords=('data', 'axes fraction'),
            xytext=(0, -3), textcoords='offset points', va='top', ha='center')

    # Adding extra features     
    plt.xlabel("Values")
    # print(ticks)
    plt.xticks(ticks='-', label='')
    axs.xaxis.set_label_coords(0.5, -0.1)
    plt.ylabel("Couns") 
    plt.legend(legend)
    plt.title(name+ ' Histogram') 
    
    # Show plot 
    plt.show()

plot_hist(df_jam_hazfi['Score Right'], 'Score Right')
plot_hist(df_jam_hazfi['Score Left'], 'Score Left')
plot_hist(df_jam_hazfi['Penalty Right'], 'Penalty Right')
plot_hist(df_jam_hazfi['Penalty Left'], 'Penalty Left')
```

```
iter_main = 0
for i in range(len(transfers.index)):
    soup_temp = BeautifulSoup(transfers.html[i], 'html.parser')
    team_list = soup_temp.find_all(class_='m3g-ct-col ct-team-name')
    team_in = soup_temp.find_all('div', {'class':'m3g-ct-col ct-in'})
    team_out = soup_temp.find_all('div', {'class':'m3g-ct-col ct-out'})

    for k in range(1, len(team_list)):
        team_name = team_list[k].get_text().strip()
        permanent_in = []
        permanent_out = []
        loan_return_in = []
        loan_return_out = []
        likely_in = []
        likely_out = []
        loan_in = []
        loan_out = []
        all_in_perm = team_in[k].find_all('span', {'class':['ct-player-name ct-permanent ', 'ct-player-name ct-permanent has-cat']})
        all_in_loan_return = team_in[k].find_all('span', {'class':['ct-player-name ct-loan-b ', 'ct-player-name ct-loan has-cat ct-loan-b']})
        all_in_likely = team_in[k].find_all('span', {'class':['ct-player-name ct-permanent likely has-cat', 'ct-player-name ct-permanent likely ']})
        all_in_loan = team_in[k].find_all('span', {'class':['ct-player-name ct-loan has-cat', 'ct-player-name ct-loan ']})

        all_out_perm = team_out[k].find_all('span', {'class':['ct-player-name ct-permanent ', 'ct-player-name ct-permanent has-cat']})
        all_out_loan_return = team_out[k].find_all('span', {'class':['ct-player-name ct-loan-b ', 'ct-player-name ct-loan has-cat ct-loan-b']})
        all_out_likely = team_out[k].find_all('span', {'class':['ct-player-name ct-permanent likely has-cat', 'ct-player-name ct-permanent likely ']})
        all_out_loan = team_out[k].find_all('span', {'class':['ct-player-name ct-loan has-cat', 'ct-player-name ct-loan ']})
        items = [all_in_perm, all_in_loan_return, all_in_likely, all_in_loan, all_out_perm, all_out_loan_return, all_out_likely, all_out_loan]
        for iteration, item in enumerate(items):
            for p in range(len(item)):
                if iteration == 0:
                    permanent_in.append(item[p].get_text().strip())
                elif iteration == 1:
                    loan_return_in.append(item[p].get_text().strip())
                elif iteration == 2:
                    likely_in.append(item[p].get_text().strip())
                elif iteration == 3:
                    loan_in.append(item[p].get_text().strip())
                elif iteration == 4:
                    permanent_out.append(item[p].get_text().strip())
                elif iteration == 5:
                    loan_return_out.append(item[p].get_text().strip())
                elif iteration == 6:
                    likely_out.append(item[p].get_text().strip())
                elif iteration == 7:
                    loan_out.append(item[p].get_text().strip())
        df_transfers.at[iter_main, 'Team'] = team_name
        year = [s for s in transfers.transfers[i].split() if s[0].isdigit()]
        transfer_name = transfers.transfers[i].replace(year[0], '')
        df_transfers.at[iter_main, 'Transfer'] = transfer_name.strip()
        df_transfers.at[iter_main, 'Year'] = year[0].strip()
        if not permanent_in:
            permanent_in = ['بدون ورودی قطعی']
        if not likely_in:
            likely_in = ['بدون ورودی احتمالی']
        if not loan_in:
            loan_in = ['بدون ورودی قرضی']
        if not loan_return_in:
            loan_return_in = ['بدون ورودی قرضی (بازگشت)']
        if not permanent_out:
            permanent_out = ['بدون خروجی قطعی']
        if not likely_out:
            likely_out = ['بدون خروجی احتمالی']
        if not loan_out:
            loan_out = ['بدون خروجی قرضی']
        if not loan_return_out:
            loan_return_out = ['بدون خروجی قرضی (بازگشت)']

        df_transfers.at[iter_main, "('ورودی', 'قطعی')"] = permanent_in
        df_transfers.at[iter_main, "('ورودی', 'احتمالی')"] = likely_in
        df_transfers.at[iter_main, "('ورودی', 'قرضی')"] = loan_in
        df_transfers.at[iter_main, "('ورودی', 'قرضی (بازگشت)')"] = loan_return_in
        df_transfers.at[iter_main, "('خروجی', 'قطعی')"] = permanent_out
        df_transfers.at[iter_main, "('خروجی', 'احتمالی')"] = likely_out
        df_transfers.at[iter_main, "('خروجی', 'قرضی')"] = loan_out
        df_transfers.at[iter_main, "('خروجی', 'قرضی (بازگشت)')"] = loan_return_out
        iter_main +=1
df_transfers.reset_index(inplace=True, drop=True)
display(df_transfers)
print(df_transfers.isna().sum())
df_transfers.to_csv("transfers_data.csv", index=False)
```

```
for i in range(len(teams.index)):
    soup_temp = BeautifulSoup(teams.html[i], 'html.parser')
    team = teams.team[i]
    table = soup_temp.find('table') #, attrs={'class':'lineItemsTable'})
    if table is not None:
        table_body = table.find('tbody')

        rows = table_body.find_all('tr')
        for row in rows:
            cols = row.find_all('td')
            cols = [ele.text.strip() for ele in cols]
            if cols[1] == team:
                df_teams.at[i, 'Playes'] = cols[2]
                df_teams.at[i, 'Scores'] = cols[3]
                break

    df_teams.at[i, 'Competition'] = teams.Competition[i]
    df_teams.at[i, 'Team'] = team
df_teams.dropna(inplace=True)
df_teams.reset_index(inplace=True, drop=True)
display(df_teams)
print(df_teams.isna().sum())
df_teams.to_csv("teams_data.csv", index=False)
```
